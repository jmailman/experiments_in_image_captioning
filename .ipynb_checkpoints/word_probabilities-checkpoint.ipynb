### Instead use Senti_WordNet

```python
import csv

def compile_senti_word_net_phrases():
    phrases_ = []
    with open('data/SentiWordNet_3.0.0.txt', 'r') as phrases_data: 
        for line in csv.reader(phrases_data):
            phrases_.extend(line)
    return phrases_[1:]
    
```

```python
len(phrases)
```

```python
phrases = compile_senti_word_net_phrases()
corpus = phrases[:50000]
```

```python
def get_related_words(df_, img_label_word):
    try:
        related_words = df_.nlargest(10,img_label_word).index.to_list()

        related_words.remove(img_label_word)
        return related_words
    except:
        return []
```
