```python
import pandas as pd
import pickle
from bs4 import BeautifulSoup
import requests
```

```python
#url = 'https://www.7esl.com'
# with open( url ) as fp:
#     soup = BeautifulSoup(fp, "lxml")

filename = 'data/1500+ English Idioms from A-Z (with Useful Examples) • 7ESL.html'
with open( filename ) as fp:
    soup = BeautifulSoup(fp, "lxml") 

idiom_list = []    
for alphabet_i in range(1, 25):
    div_id = 'elementor-tab-content-240' + str(alphabet_i)
    alphabetic_category = soup.find('div', id=div_id)
    idiom_packages_for_alpha_letter = alphabetic_category.find('table').find_all('strong')
    for idiom_package in idiom_packages_for_alpha_letter:
        #print(idiom_package.text)
        idiom_list.append(idiom_package.text)
print( idiom_list )
```
